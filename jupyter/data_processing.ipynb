```python
from sklearn.ensemble import RandomForestRegressor
```

```python
base_train = pd.read_csv("./base_train.csv")
```

```python
rf_test = df_data[df_data["feature_8"].isnull().values == True]
```

```python
cols = list(rf_test.columns)
print(cols)
cols_FT = list(rf_test.isnull().any())
[k for k,v in zip(cols, cols_FT) if v == False]
```

```python
df_nan_data = pd.read_csv('./train_with_nan.csv',encoding="gbk")
c = "feature_0"
rf_pre_data = df_nan_data[df_nan_data[c].isnull().values == True]
no_index = [x for x in list(df_nan_data.index) if x not in list(rf_pre_data.index)]
rf_no_pre_data = df_nan_data.ix[no_index,:]
all_rf_data = pd.concat([rf_no_pre_data, rf_pre_data],axis=0)
no_nan_cols = [k for k,v in zip(list(rf_pre_data.columns),list(rf_pre_data.isnull().any())) if
               v == False and k not in ['id',"date","blood_sugar"]]
mean = round(df_nan_data[c].mean(),3)
trains = np.array(base_train[no_nan_cols])
labels = np.array(base_train[c])
tests = np.array(rf_pre_data[no_nan_cols])
rf = RandomForestRegressor()
rf.fit(trains, labels)
rf_pre_data[c] = [(2 * x * mean)/ (x + mean) for x in list(rf.predict(tests))]
df_nan_data.ix[rf_pre_data.index,c] = rf_pre_data[c]
```

```python
df_nan_data = pd.read_csv('./train_with_nan.csv',encoding="gbk")
print(round(df_nan_data[c].mean(),3))
print(df_nan_data.ix[rf_pre_data.index,c])
df_nan_data.ix[rf_pre_data.index,c] = rf_pre_data[c]
print(rf_pre_data[c])
print(df_nan_data.ix[rf_pre_data.index,c])
print(df_nan_data[c])
```
