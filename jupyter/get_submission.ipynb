# 导入相关的包

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
pred1 = pd.read_csv("../submission/biorad_sub_lgb_c_8179.csv",names=["pred"])
pred2 = pd.read_csv("../raw_data/d_answer_a_20180128.csv",names=["pred"])
res = pd.DataFrame()
res["pred1"] = pred1.pred.values
res["pred2"] = pred2.pred.values
res["pred"] = res["pred2"] - res["pred1"]
```

```python
res.to_csv("../bi.csv",index=False)
```

```python
pred1 = pd.read_csv("../submission/lgb.csv",names=["pred"])
pred2 = pd.read_csv("../submission/df_lgb.csv",names=["pred"])
res = pd.DataFrame()
res["pred1"] = pred1.pred.values
res["pred2"] = pred2.pred.values
res["pred"] = (res["pred1"] + res["pred2"]) / 2
res["pred"] = round(res["pred"],3)
res["pred"].to_csv("../submission/biorad.csv",index=False, header=False)
```

```python
res.sort_values(by="pred",ascending=False)
```

# 导入相关数据

```python
old = pd.read_csv("../regression/sub_B.csv")
result_pred = pd.read_csv("../submission/sub_lgb_B_28_h.csv",names=["pred"])
result_pred["id"] = old["id"]
print("begin process gt10..")
```

```python
result_pred.head()
```

```python
old.head()
```

```python
gt10_prob = pd.read_csv("../classification/result/gt10.csv")
print(gt10_prob.index)
gt10_prob.sort_values(by='gt10_prob',inplace=True)
gt10_prob = gt10_prob.tail(9)
result_pred = pd.merge(result_pred,gt10_prob,on='id',how='left')
result_pred.fillna(-999,inplace=True)
result_pred_sou = result_pred[result_pred.gt10_prob==-999]
result_pred_gt = result_pred[result_pred.gt10_prob != -999]
result_pred = result_pred_sou[['id','pred']]
result_pred_gt["pred"] = result_pred_gt["pred"] * 1.2
result_pred_gt = result_pred_gt[["id","pred"]]
```

```python
gt10_prob
```

```python
result_pred_gt
```

```python
lt45_prob = pd.read_csv('../classification/result/lt45.csv')
lt45_prob.sort_values(by='lt45_prob',inplace=True)
lt45 = lt45_prob.tail(5)
result_pred = pd.merge(result_pred,lt45,on='id',how='left')
result_pred.fillna(-999,inplace=True)
result_pred_sor = result_pred[result_pred.lt45_prob==-999]
result_pred_lt = result_pred[result_pred.lt45_prob !=-999]
result_pred = result_pred_sor[['id','pred']]
result_pred_lt["pred"] = result_pred_lt["pred"] * 0.9
result_pred_lt = result_pred_lt[["id","pred"]]
```

```python
submission = pd.concat([result_pred,result_pred_lt,result_pred_gt],axis=0)
submission.sort_values(by='id',inplace=True)
submission["pred"].to_csv('../submission/final_predict_result_stacking_21_2.csv',index=False, header=False)
```
