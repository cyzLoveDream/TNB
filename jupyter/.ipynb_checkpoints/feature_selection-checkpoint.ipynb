# 导入相关的包

```python
import pandas as pd
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

# 导入数据

```python
%time
trains = pd.read_csv('../clean_data/base_train.csv',encoding="gbk")
labels = "blood_sugar"
features = [x for x in list(trains.columns) if x not in ['id','date','blood_sugar','blood_sugar_log']]
features_with_no_cate = [x for x in list(trains.columns) if x not in ['id','date','blood_sugar_log','gender','age']]
vals = pd.read_csv('../clean_data/model_fill_train.csv',encoding='gbk')
trains = pd.concat([trains, vals],axis=0)
train = trains[features_with_no_cate]
y_train = trains[labels]
gender_onehot_train = pd.get_dummies(trains['gender'])
gender_onehot_train.rename(columns={0:'gender_0',1:'gender_1'},inplace=True)
train = pd.concat([train, gender_onehot_train, trains['age']],axis=1)
```

```python
train.shape
train.columns
```

```python
data = pd.read_csv("../raw_data/d_train.csv",encoding="gbk")
rename_dict = {"性别":"gender","年龄":"age","体检日期":"date","血糖":"blood_sugar"}
data.rename(columns=rename_dict,inplace=True)
features = [x for x in list(data.columns) if x not in ["gender","age","date","blood_sugar","id"]]
fea_rename = { "feature_" + str(k):v for k,v in enumerate(features)}
fea_rename
```

# 特征选择模型

## 互信息和最大信息系数

```python
from minepy import MINE
m = MINE()
m_list = []
m_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
for i in train.columns :
    if i in m_cols:
        m_list.append([i, m.compute_score(train[i],train['blood_sugar'])])
    else:
        m_list.append([fea_rename.get(i), m.compute_score(train[i],train['blood_sugar'])])
```

## 随机森林回归

```python
from sklearn.cross_validation import cross_val_score, ShuffleSplit
from sklearn.ensemble import RandomForestRegressor
rf_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
rf_features = list(train.columns)
rf_features.remove("blood_sugar")
X = np.array(train[rf_features])
Y = np.array(train["blood_sugar"])
rf = RandomForestRegressor(n_estimators=20, max_depth=4)
scores = []
for i in range(X.shape[1]):
    score = cross_val_score(rf, X[:, i:i+1], Y, scoring="r2",
                              cv=ShuffleSplit(len(X), 3, .3))
    if rf_features[i] not in rf_cols:
        scores.append((round(np.mean(score), 3), fea_rename.get(rf_features[i])))
    else:
        scores.append((round(np.mean(score), 3), rf_features[i]))
print(sorted(scores, reverse=True))
```


### 平均不纯度减少 mean decrease impurity

```python
from sklearn.cross_validation import cross_val_score, ShuffleSplit
from sklearn.ensemble import RandomForestRegressor
rf_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
rf_features = list(train.columns)
rf_features.remove("blood_sugar")
X = np.array(train[rf_features])
Y = np.array(train["blood_sugar"])
rf = RandomForestRegressor(random_state=530)
rf.fit(X, Y)
print("Features sorted by their score:")
re_rf_features = []
for x in rf_features:
    if x in rf_cols:
        re_rf_features.append(x)
    else:
        re_rf_features.append(fea_rename.get(x))
print(re_rf_features)
print(rf.feature_importances_)
print(sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), re_rf_features), 
             reverse=True))
```

### 平均精确率减少 Mean decrease accuracy

```python
from sklearn.cross_validation import cross_val_score, ShuffleSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from collections import defaultdict
rf_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
rf_features = list(train.columns)
rf_features.remove("blood_sugar")
X = np.array(train[rf_features])
Y = np.array(train["blood_sugar"])
rf = RandomForestRegressor(random_state=530)
rf.fit(X, Y)
print("Features sorted by their score:")
re_rf_features = []
for x in rf_features:
    if x in rf_cols:
        re_rf_features.append(x)
    else:
        re_rf_features.append(fea_rename.get(x))
scores = defaultdict(list)
#crossvalidate the scores on a number of different random splits of the data
for train_idx, test_idx in ShuffleSplit(len(X), 500, .3):
    X_train, X_test = X[train_idx], X[test_idx]
    Y_train, Y_test = Y[train_idx], Y[test_idx]
    r = rf.fit(X_train, Y_train)
    acc = r2_score(Y_test, rf.predict(X_test))
    for i in range(X.shape[1]):
        X_t = X_test.copy()
        np.random.shuffle(X_t[:, i])
        shuff_acc = r2_score(Y_test, rf.predict(X_t))
        if rf_features[i] not in rf_cols:
            scores[fea_rename.get(rf_features[i])].append((acc-shuff_acc)/acc)
        else:
            scores[rf_features[i]].append((acc-shuff_acc)/acc)
print("Features sorted by their score:")
print(sorted([(round(np.mean(score), 4), feat) for feat, score in scores.items()], reverse=True))
```

## 线性模型和正则化

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
lr_features = list(train.columns)
lr_features.remove("blood_sugar")
X = np.array(train[lr_features])
Y = np.array(train["blood_sugar"])
lr.fit(X, Y)
#A helper method for pretty-printing linear models
def pretty_print_linear(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
print("Linear model:", pretty_print_linear(lr.coef_))
```

### LASSO

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import ElasticNet, Lasso,  BayesianRidge, LassoLarsIC
from sklearn.preprocessing import RobustScaler
# lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005, random_state=42))
la_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
la_features = list(train.columns)
la_features.remove("blood_sugar")
X = np.array(train[la_features])
Y = np.array(train["blood_sugar"])
rs = RobustScaler()
X = rs.fit_transform(X)
#A helper method for pretty-printing linear models
def pretty_print_linear(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    la_names = []
    for n in names:
        if n in la_cols:
            la_names.append(n)
        else:
            la_names.append(fea_rename.get(n))
    lst = zip(coefs, la_names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
l = Lasso(alpha =0.0005, random_state=42).fit(X, Y)
print("Lasso model: ", pretty_print_linear(l.coef_, la_features, sort = True))
```

## 基于模型的特征选择

```python
from sklearn.linear_model import RandomizedLasso
rl_cols = ["feature_15","feature_16","feature_17","feature_18","feature_19","age","gender_0","gender_1"]
rl_features = list(train.columns)
rl_features.remove("blood_sugar")
X = np.array(train[rl_features])
Y = np.array(train["blood_sugar"])
rl = RandomizedLasso(alpha=0.025)
rl.fit(X, Y)
re_rf_features = []
for x in rf_features:
    if x in rf_cols:
        re_rf_features.append(x)
    else:
        re_rf_features.append(fea_rename.get(x))
print("Features sorted by their score:")
print(sorted(zip(map(lambda x: round(x, 4), rl.scores_), 
                 re_rf_features), reverse=True))
```
